# vii) get the list of hospital which is for senior citizen and child care

```python

# Get hospitals for senior citizens
senior_hospitals = ho[ho['Age'].str.contains('6|7|8|9|10')]

# Get hospitals for children
child_hospitals = ho[ho['Age'].str.contains('0|1|2|3|4|5')]

# Combine and remove duplicates
combined_hospitals = pd.concat([senior_hospitals, child_hospitals]).drop_duplicates()

# Display the results
combined_hospitals[['Hospital Name', 'Address','Age', 'City', 'State', 'ZIP_Code']]
```
